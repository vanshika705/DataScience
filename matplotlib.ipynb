<h2>LINE PLOT</h2>
<p>1. ms ---> marker size<br>
 2. mfc ---> marker face color<br> 
 3. show() ---> does not show other info <br>
 4. line plot is used to check trend <br>

 6. ls ---> line style<br>
 7. lw ---> line width <br>
 8. values should be equal in both list <p>

<h2>SCATTER GRAPH</h2>

2. x and y must be numerical for scatter plot<br>
3. col should be 2 and must be numerical<br>
<h3>TYPES OF RELATION :-</h3><br> 1. +ve relation ---> x and y both increases simultaneously <br> 2. -ve relation ---> inverse relationship btw x and y </p>

```python
royal = df[(df["brand"]=="Royal Enfield")&(df["price"]<=50000)]
price = list(royal["price"].sort_values())
age = list(royal["age"].sort_values())
pt = plt.plot(price,age,marker =".",ms=12)
plt.xlabel("PRICE")
plt.ylabel("AGE")
```

<h2>BAR GRAPH</h2>
<p>1. Bar Graph is used when one col is catgorical and other col is numerical<br>
    </p>

```python
brand=list(df.brand.value_counts().keys())                #categorical
no_of_count=list(df.brand.value_counts().values)          #numerical
brand
```

```python
no_of_count
```

```python
plt.bar(brand,no_of_count,color="green",ec="k",lw=4,ls="solid",width=0.8, )  
# plt.figure(figsize=(16,8))  #(width,height)
plt.xlabel("brand")
plt.ylabel("no_of_count")
plt.title('no. of bikes in brand')
plt.xticks(rotation=90)
plt.show()
```
