```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Used_Bikes.csv")
df
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df["brand"].value_counts()>=10
```

```python
df.brand.value_counts().keys()
```

```python
brand_list=['Bajaj', 'Royal Enfield', 'Hero', 'Honda', 'Yamaha', 'TVS', 'KTM',
       'Suzuki', 'Harley-Davidson', 'Kawasaki', 'Hyosung', 'Mahindra',
       'Benelli', 'Triumph', 'Ducati', 'BMW']
```

```python
filter_brand=df[df.brand.isin(brand_list)]
filter_brand
```

```python
cat=df.select_dtypes(include="O")
cat
```

```python
num=df.select_dtypes(exclude="O")
num
```

```python
cat["owner"].value_counts()
```

```python
dt={"First Owner":1, "Second Owner ":2,"Third Owner":3,"Fourth Owner Or More":4}
cat["owner"]=cat["owner"].map(dt)
cat["owner"]
```

```python
brand_list
```

```python
dt2={item:i for i , item in enumerate(brand_list,start=1)}
dt2
```

```python
cat["brand"]=cat["brand"].map(dt2)
cat["brand"]
```

```python
cat.drop(["bike_name","owner","city"],axis="columns",inplace=True)
```

```python
combine=pd.concat([cat,num],axis=1)
combine
```

```python
combine.dropna(inplace= True)
```

```python
x=combine.drop("price",axis=1)
x
```

```python
y=combine[["price"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lg=LinearRegression()
```

```python
x_train.isnull().sum()
```

```python
lg.fit(x_train,y_train)
```

```python
print("Training score :",round(lg.score(x_train,y_train)*100,2))
print("Testing score :",round(lg.score(x_test,y_test)*100,2))
```

```python
y_hat=lg.predict(x_test)
```

```python
y_test["prediction"]=y_hat
```

```python
y_test["prediction"]=y_test["prediction"].map(lambda x:round(x,2))
```

```python
from sklearn.ensemble import RandomForestRegressor
rdf=RandomForestRegressor()
```

```python
rdf.fit(x_train,y_train)
```

```python
print("Training score :",round(rdf.score(x_train,y_train)*100,2))
print("Testing score :",round(rdf.score(x_test,y_test["price"])*100,2))
```
