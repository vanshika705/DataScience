```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Used_Bikes.csv")
df
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df["brand"].value_counts()>=10
```

```python
df.brand.value_counts().keys()
brand_list=['Bajaj', 'Royal Enfield', 'Hero', 'Honda', 'Yamaha', 'TVS', 'KTM',
       'Suzuki', 'Harley-Davidson', 'Kawasaki', 'Hyosung', 'Mahindra',
       'Benelli', 'Triumph', 'Ducati', 'BMW']
```

```python
filter_brand=df[df["brand"].isin(brand_list)]
filter_brand
```

```python
# y=price , x=other columns 
# ML predicts price (y) on the basis of other columns (x).
# Price is in continuous nature i.e Regression.
```

```python
categorical=df.select_dtypes(include="O")
categorical
```

```python
numerical=df.select_dtypes(exclude="O")
numerical
```

```python
# ML algorithm cannot accept alphabetical data.
categorical["owner"].value_counts()
```

```python
dt={"First Owner" :1,"Second Owner":2,"Third Owner" :3,"Fourth Owner Or More":4 }
categorical["owner"]=categorical["owner"].map(dt)
```

```python
brand_list
```

```python
dt2= {item:i for i,item in enumerate(brand_list,start=1)}
dt2
```

```python
categorical["brand"]=categorical["brand"].map(dt2)
categorical["brand"]
```

```python
categorical.drop(["bike_name","owner","city"],axis="columns",inplace=True)
```

```python
categorical
```

```python
combined=pd.concat([categorical,numerical],axis=1)
combined
```

```python
combined.dropna(inplace=True)
```

```python
# arrange data
x=combined.drop("price",axis=1)
x
```

```python
y=combined[["price"]]
y
```

```python
# train test split 
# divide rows randomly in two parts train and test,ex= 80% and 20% of 1000 rows
# x_train , y_train ,x_test, y_test
# 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train  ,x_test,y_train, y_test=train_test_split(x,y,test_size=0.2)
```

```python
# y_hat prediction is prediction given by algo
# in testing only x is given 
# in training x and y both are given 
#  
```

```python

from sklearn.linear_model import LinearRegression
```

```python
# object of LinearRegression class
lg=LinearRegression()
```

```python
x_train.isnull().sum()
```

```python
x_train.head()
```

```python
# linear regression algo training
# training data for training purpose
lg.fit(x_train,y_train)
```

```python
# test model= model evaluation 
# model training  , measure the performance of model
# classification and regression both have diff method of evaluation
# 
```

```python
# training and testing
print("Training score :",round(lg.score(x_train,y_train)*100,2))
print("Testing score :",round(lg.score(x_test,y_test)*100,2))
```

```python
# in above model 
```

```python
# accuracy should be greater than 90
# diff btw training and testing score should not more than 5
# Is model overfit or underfit or perfect(generalized model)
# if model overfit = trainig score ---> high , testing score ---> low
# if model underfit = trainig score ---> low , testing score ---> high/low
# if generalized model = diff not more than 5 , model should not be overfit or underfit,score and accuracy should be high in both testing and training 

```

```python
# prediction --> pass x data to algo
y_hat=lg.predict(x_test)
```

```python
x_test.shape
```

```python
y_test["prediction"]=y_hat
```

```python
y_test["prediction"]=y_test["prediction"].map(lambda x:round(x,2))
```

```python
y_test.head(40)
```

```python
from sklearn.ensemble import RandomForestRegressor
rdf=RandomForestRegressor()
```

```python
rdf.fit(x_train,y_train)
```

```python
print("Training score :",round(rdf.score(x_train,y_train)*100,2))
print("Testing score :",round(rdf.score(x_test,y_test["price"])*100,2))
```
