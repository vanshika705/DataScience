```python
df=pd.read_csv("Used_Bikes.csv")
df.head()
```

```python
ktm=df[df.brand=="KTM"]
ktm
```

```python
price=list(ktm["price"].values)
age=list(ktm["age"].values)
age.sort()
price.sort()
```

```python
df.drop_duplicates("price",inplace=True)
```

```python
mp.plot(df["price"])
```

```python
mp.plot(age,price)
```

```python
arr=np.random.randint(1,50,20)
mp.plot(arr,color="pink",lw=4)
```

```python
price=list(df["price"].values)
age=list(df["age"].values)
mp.scatter(age,price)
mp.xlabel("price")
mp.ylabel("age")
mp.title("price vs age ")
mp.show()
```

```python
brands=list(df.brand.value_counts().keys())          #categorical
brand_count=list(df.brand.value_counts().values)     #numerical
mp.bar(brands,brand_count,color="yellow",ec="r",lw=2,ls="solid",width=0.8)
mp.xlabel("brand")
mp.ylabel("brand_count")
mp.title("brands vs brand_count")
mp.xticks(rotation=90)
mp.show()
```

```python
ar=np.random.randint(1,100,50)
bin=[0,15,37,69,73,82,97]
mp.hist(ar,bins=bin,color="pink",ec="aqua",lw=4)
mp.xlabel("age")
mp.ylabel("number of people")
mp.title("Age Chart")
mp.show()
```

```python
bike_age=list(df.age.values)
bin=[0,3,5,8,11,13,18,21,25]
mp.hist(bike_age,bins=bin,color="orange",ec="purple",lw=4)
mp.xlabel("bike age")
mp.ylabel("km runs")
mp.title("Bike Chart")
mp.show()
```
