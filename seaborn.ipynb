<h1>SEABORN</h1>

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df= sns.load_dataset("tips")
df.head()
```

```python
df.columns=['total_bill', 'tip', 'gender', 'smoker', 'day', 'time', 'size']
```

<h2>TYPES OF PLOT IN SEABORN :-</H2>
<p>relplot() ---> relational plot are used when  all columns are numerical. <br>
Relational plot are scatter plot by default <br>
catplot() ---> catgorical plot are used when one column is numerical and other is catgorical.<br>
kind is used to change scatter graph into line graph.<br> 
These are used to create any type of  charts :-<br>
1. sns.lineplot()<br>
2. sns.barplot()<br>
3. sns.scatterplot()<br>
4. sns.boxplot()<br>
5. sns.countplot()<br>
6. sns.histplot() </p>

```python
sns.scatterplot(x="total_bill",y="tip",data=df)
```

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="gender",style="smoker",)
```

<p>1. hue ---> gives color to different values of a column<br>
2. col ---> make different graphs for different values of a column and stack them horizontally <br>
3. style ---> gives different shape to the marker of the given column<br>
4. row --->  make different graphs for different values of a column and stack them vertically <br></p>


```python
sns.relplot(x="total_bill",y="tip",data=df,hue="gender",style="smoker",col="time")
```

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="gender",col="day")
```

```python
female =df[df["gender"]=="Female"]
```

```python
sns.relplot(x="tip",y="total_bill",data=female,hue="smoker",row="day")
```
