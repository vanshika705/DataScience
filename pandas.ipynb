<h1>11. Mapping of new values in a column.</h1>
<p>To transform categorial values into numerical values<p> 

```python
label={"First Owner":1,          
        "Second Owner " :2 ,          
        "Third Owner ":  3 ,          
        "Fourth Owner Or More " :4    }
label
```

```python
# mapping of numerical values in the place or categorial place 
output=df["owner"].map(label)
output
```

```python
df["owner"]=output
df
```
